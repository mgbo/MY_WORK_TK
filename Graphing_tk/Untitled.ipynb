```python
import time
import psutil
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
# plt.rcParams['animation.html'] = 'jshtml'
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
```

```python
psutil.cpu_percent()
```

```python
i = 0
x,y =[], []

while True:
    x.append(i)
    y.append(psutil.cpu_percent())
    
    ax.plot(x, y, color='b')
    
    fig.canvas.draw() #á€•á€¯á€¶á€€á€­á€¯ update á€•á€¼á€¯á€œá€¯á€•á€ºá€›á€”á€ºá€¡á€á€½á€€á€º
    ax.set_xlim(left=max(0, i-50), right=i+50)
    time.sleep(0.1)# 0.1 á€…á€€á€¹á€€á€”á€·á€º
    i +=1
```

```python
plt.close()
```
